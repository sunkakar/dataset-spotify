```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import spotipy.util as util
import json
import csv
```

```python
# Spotify Credentials
client_id = 'XXX'
client_secret = 'XXX'

# Spotify Authentication via spotipy
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# Collecting new columns on the data rows
def get_track(track_id):
    track_info = sp.track(track_id)
    artists = ", ".join([artist['name'] for artist in track_info['artists']])
    album = track_info['album']['name']
    popularity = track_info['popularity']
    duration_ms = track_info['duration_ms']
    explicit = track_info['explicit']
    release_date = track_info['album']['release_date']
    preview_url = track_info['preview_url']
    external_urls = track_info['external_urls']
    available_markets = track_info['available_markets']
    audio_features = sp.audio_features([track_id])[0]

    return (artists, album, popularity, duration_ms, explicit, release_date,
            preview_url, external_urls, available_markets, audio_features)
```

```python
# Example run
track_id = '6mICuAdrwEjh6Y6lroV2Kg'
result = get_track(track_id)
print("Result:")
print(json.dumps(result, indent=2))
```

```python
# Updating gettrack to only contain new columns and restructure
def notNone(s,d):
    if s is None:
        return d
    else:
        return s

def get_track_info(track_id):

    returndata = []
    trackdata = sp.tracks(track_id)
    afdata = sp.audio_features(track_id)

    print("Track Data:", len(trackdata['tracks']), trackdata)
    print("Audio Features:", len(afdata), afdata)

    for id in range(0,len(trackdata['tracks'])):
      # print("Print",track_info)

      track_info = trackdata['tracks'][id]
      audio_features = afdata[id]
      if(audio_features is None):
          continue

      album = track_info['album']['name']
      popularity = track_info['popularity']
      duration_ms = track_info['duration_ms']
      explicit = track_info['explicit']
      release_date = track_info['album']['release_date']
      available_markets = track_info['available_markets']
      af_danceability = audio_features["danceability"] if audio_features is not None else ''
      af_energy = audio_features["energy"]
      af_key = audio_features["key"]
      af_loudness = audio_features["loudness"]
      af_mode = audio_features["mode"]
      af_speechiness = audio_features["speechiness"]
      af_acousticness = audio_features["acousticness"]
      af_instrumentalness = audio_features["instrumentalness"]
      af_liveness = audio_features["liveness"]
      af_valence = audio_features["valence"]
      af_tempo = audio_features["tempo"]
      af_time_signature = audio_features["time_signature"]

      returndata.append({
          "track_id": track_id[id],
          
          "album": album,
          "popularity": popularity,
          "duration_ms": duration_ms,
          "explicit": explicit,
          "release_date": release_date,
          "available_markets": available_markets,
          "af_danceability": af_danceability,
          "af_energy": af_energy,
          "af_key": af_key,
          "af_loudness": af_loudness,
          "af_mode": af_mode,
          "af_speechiness": af_speechiness,
          "af_acousticness": af_acousticness,
          "af_instrumentalness": af_instrumentalness,
          "af_liveness": af_liveness,
          "af_valence": af_valence,
          "af_tempo": af_tempo,
          "af_time_signature": af_time_signature
      })

    return returndata

track_id = ['6mICuAdrwEjh6Y6lroV2Kg','1AkTW13ysu0AJrwuM6UY0I']
result = get_track_info(track_id)
print("Result:")
print(json.dumps(result, indent=2))
```
