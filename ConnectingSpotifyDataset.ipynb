```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import spotipy.util as util
import json
import csv
```

```python
# Load data from CSV
data = []

for chunk in pd.read_csv('spotify-charts/charts.csv', chunksize=1000000):
  print('Chunk', chunk.shape)
  data.append(chunk)

spdata = pd.concat(data)
spdata.head()
```

```python
# Load data from CSV
data = []

for chunk in pd.read_csv('spotify_data.csv', chunksize=1000000):
  print('Chunk', chunk.shape)
  data.append(chunk)

newdata = pd.concat(data)
newdata.head()
```

```python
# connect the tables

newdata['url'] = 'https://open.spotify.com/track/' + newdata['track_id']

merged_data = pd.merge(spdata, newdata, on='url', how='left')
```

```python
merged_data.to_csv('merged_data.csv')
```
